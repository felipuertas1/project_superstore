# **FEATURE SELECTION**

La matriz de correlacion fue la feature selection elegida con el objetivo de encontras las relaciones entre las variables y reducirlas para quedarte solo con las mas relevantes del modelo.

```python
df_num = df_superstore.select_dtypes(include=['int64', 'float64'])
df_num = df_superstore.drop(columns=['Row ID', 'Postal Code'], errors='ignore')
```

```python

corr_matrix = df_num.corr(numeric_only=True)

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Matriz de correlación")
plt.show()

umbral = 0.85

correlated_features = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > umbral:
            colname = corr_matrix.columns[i]
            correlated_features.add(colname)

df_reducido = df_num.drop(columns=correlated_features)

print("Columnas eliminadas por alta correlación:", correlated_features)
print("Forma original:", df_superstore.shape)
print("Forma reducida:", df_reducido.shape)
```




Los resultados obtenidos muestran que no existen correlaciones fuertes entre la mayoría de las variables analizadas, lo que indica una baja dependencia lineal entre ellas. Sin embargo, se destaca la relación entre “Profit” (ganancia) y “Sales” (ventas), con un coeficiente de correlación de 0.48, lo que representa una asociación positiva moderada: a medida que aumentan las ventas, las ganancias tienden a incrementarse, aunque no de manera proporcional o completamente lineal.

El resto de las variables, como “Quantity” y “Discount”, presentan correlaciones muy bajas con las demás, lo que sugiere que su impacto directo sobre las ganancias o las ventas es limitado. En particular, la ligera correlación negativa entre “Discount” y “Profit” (-0.22) refuerza la idea de que mayores descuentos pueden reducir levemente la rentabilidad.

En conclusión, la matriz de correlación permite identificar que “Sales” es la variable más relevante para el análisis del “Profit”, mientras que las demás variables no muestran relaciones lineales significativas.

# **ALGORITMO DE CLASIFICACION**

Para este paso elegi el metodo k-means que seria un modelo no supervisado.

Fue seleccionado por su capacidad para identificar patrones y agrupar observaciones con características similares, sin necesidad de contar con una variable objetivo.

Este algoritmo de aprendizaje no supervisado me permite segmentar el dataset según variables clave como Sales, Profit, Quantity y Discount, facilitando la detección de comportamientos y tendencias que no son evidentes a simple vista.

Aplicar el K-Means me resulta para analizar la rentabilidad y eficiencia de las operaciones comerciales, ayudando a distinguir distintos perfiles de ventas y a obtener conclusiones que aportan valor al negocio.

```python
from sklearn.cluster import KMeans

df = df_superstore.copy()

cols_to_drop = ['Customer Name', 'Product Name', 'Postal Code', 'Country', 'Order ID', 'Order Date', 'Ship Date', 'Row ID','Region','Segment','Ship Mode']
df = df.drop(columns=cols_to_drop, errors='ignore')

categorical_cols = ['Category', 'Sub-Category']

df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

X = df_encoded[['Sales', 'Quantity', 'Discount']]

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_encoded['Cluster'] = kmeans.fit_predict(X)

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Sales', y='Quantity', hue='Cluster', data=df_encoded, palette='Set2')
plt.title('Clusters detectados por K-Means (Ventas vs Cantidad de Ventas)')
plt.xlabel('Ventas')
plt.ylabel('Cantidad de ventas')
plt.show()

print("\nPromedio de variables por cluster:")
print(df_encoded.groupby('Cluster')[['Sales', 'Quantity', 'Discount']].mean())
```

En este grafico se muestran los clusteres detectados por el algoritmo K-Means, utilizando las variables VENTAS (Sales) y Cantidad de Ventas (Quantity).
Cada punto representa una venta expuesta en el dataset y los colores indican a que cluster pertenece segun las similitudes encontradas por el modelo.

El objetivo del analisis fue identificar patrones o segmentos de comportamiento entre las ventas y las cantidades vendidas.

```python

df = df_superstore.copy()

cols_to_drop = [
    'Customer Name', 'Product Name', 'Postal Code', 'Country',
    'Order ID', 'Order Date', 'Ship Date', 'Row ID',
    'Region', 'Segment', 'Ship Mode'
]
df = df.drop(columns=cols_to_drop, errors='ignore')

categorical_cols = ['Category', 'Sub-Category']
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)


X = df_encoded[['Sales', 'Quantity', 'Discount']]


kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_encoded['Cluster'] = kmeans.fit_predict(X)


plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df_encoded,
    x='Sales',
    y='Quantity',
    hue='Cluster',
    palette={0: "#7E77FF", 1: "#FF9A76", 2: "#36CFC9"},
    s=70,
    alpha=0.8
)


centroids = df_encoded.groupby('Cluster')[['Sales', 'Quantity']].mean().values
plt.scatter(
    centroids[:, 0], centroids[:, 1],
    c='black',
    marker='X',
    s=200,
    label='Centroides'
)

plt.title('Clusters detectados por K-Means (Ventas vs Cantidad de Ventas)', fontsize=14, fontweight='bold')
plt.xlabel('Ventas', fontsize=12)
plt.ylabel('Cantidad Vendida', fontsize=12)
plt.legend(title='Cluster', loc='upper left')
plt.tight_layout()
plt.show()

print("\nPromedio de variables por cluster:")
promedios = df_encoded.groupby('Cluster')[['Sales', 'Quantity', 'Discount']].mean().round(2)
print(promedios)
```

A diferencia del anterior grafico, en este se muestran los tres clusters representados por los 3 centroides (marcados por una X Negra) que representan el promedio de cada grupo.

El cluster 0 (color Azul): agrupa trasacciones pequeñas, con bajas ventas y menor cantidad de productos vendidos. Es el grupo de ventas menores o de bajo volumen.

El cluster 1 (color naranja claro): Serian las ventas medias, con una cantidad vendida algo superior. nivel intermedio de ventas

El cluster 2 (color celeste): Corresponde a ventas altas, con grandes montos y cantidades moderadas, y con el menor nivel de descuentos promedio, lo que sugiere clientes o productos de alto valor.

# **CALCULO DE METRICAS PARA VALIDAR EL MODELO**

Para evaluar el algoritmo se utilizaron las siguientes metricas de clustering con el fin de validar la "cohesion" y separacion de los grupos formados:



1- Inercia (WCSS): ¿Que hace? = mide la compacidad interna de los clusters.

```python
print("Inercia del modelo:", kmeans.inertia_)
```

La inercia del modelo K-Means fue de 1.016.855.412, lo que representa la suma total de las distancias cuadradas entre los puntos y los centroides de sus respectivos clusters.
Este valor refleja la cohesión interna de los grupos: cuanto menor sea la inercia, más compactos y homogéneos son los clusters.
En este caso, el valor obtenido indica una dispersión moderada, coherente con la variabilidad propia de las variables analizadas (Profit, Sales, Quantity, Discount).

```python
from sklearn.metrics import silhouette_score

silhouette = silhouette_score(X, kmeans.labels_)
print("Coeficiente de silueta:", silhouette)
```

El resultado obtenido por coeficiente de silueta de 0.8567, indica que exite una excelente calidad de agrupamiento.
Este valor sugiere que los puntos dentro de cada cluster son muy similares entre sí (alta cohesión), mientras que los distintos clusters están bien diferenciados (buena separación).
¿Que genera esto? que el modelo K-Means logró una segmentación clara y coherente con los patrones del dataset.

```python
from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score

ch = calinski_harabasz_score(X, kmeans.labels_)
db = davies_bouldin_score(X, kmeans.labels_)

print("Calinski-Harabasz:", ch)
print("Davies-Bouldin:", db)
```


El índice de Calinski–Harabasz (14.073,98) presentó un valor elevado, lo cual indica una buena separación entre los clusters y una cohesión interna adecuada.

Asimismo, el índice de Davies–Bouldin (0.47) fue bajo, lo que refuerza la idea de que los grupos obtenidos están bien definidos y no presentan superposición significativa.
